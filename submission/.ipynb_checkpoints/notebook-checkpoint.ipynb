```python
Visualization & Explanatory Analysis
Bagaimana pengaruh musim terhadap penyewaan sepeda?
```

```python
byseason_df = day_df.groupby(by="season").instant.nunique().reset_index()
byseason_df.rename(columns={
    "season": "musim",
    "instant": "jumlah"
}, inplace=True)
byseason_df
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    y="jumlah",
    x="musim",
    data=byseason_df.sort_values(by="musim", ascending=False)
)
plt.title("Jumlah Bike Sharing Berdasarkan Musim", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis="x", labelsize=10)
plt.show()
```

```python
Bagaimana tren penyewaan sepeda dalam tahun, bulan dan jam?
```

```python
byyear_df = day_df.groupby(by="yr").instant.nunique().reset_index()
byyear_df.rename(columns={
    "yr": "tahun",
    "instant": "jumlah"
}, inplace=True)
byyear_df
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    y="jumlah",
    x="tahun",
    data=byyear_df.sort_values(by="tahun", ascending=False),
)
plt.title("Jumlah Sepeda Sharing Berdasarkan Tahun", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis="x", labelsize=10)
plt.show()
```
